```python
plt.figure(figsize=(10,5))
sns.histplot(df["track_popularity"], kde=True)
plt.title("Track Popularity Histogram")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df["track_duration_min"], kde=True)
plt.title("Track Duration")
plt.show()
```

```python
df["explicit"].value_counts().plot(kind="pie", autopct = '%1.1f%%', figsize=(4,4))
plt.title("Explicit and Non-Explicit Tracks")
plt.ylabel(" ")
plt.show()
```

```python
top_artists = df["artist_name"].value_counts().head(10)
top_artists.plot(kind='pie', autopct = '%1.1f%%', figsize=(10,5))
plt.title("Top 10 Artists by Number of Tracks")
plt.ylabel(" ")
plt.show()

top_artists.plot(kind='bar', figsize=(10,5), color= 'purple')
plt.xlabel("Artist Name")
plt.ylabel("Number Of Tracks")
plt.show()
```

```python
top_genres = df["artist_genres"].value_counts().head(5)
top_genres.plot(kind="bar", figsize=(12,5), color = 'teal')
plt.xlabel("Top 5 Genres")
plt.ylabel(" ")
plt.show()
```

```python
df["year"]= pd.to_datetime(df["album_release_date"]).dt.year
plt.figure(figsize=(12,6))
df["year"].value_counts().sort_index().plot(kind="line", marker="o", color = "darkblue")
plt.title("Tracks Released per Year")
plt.show()
```
